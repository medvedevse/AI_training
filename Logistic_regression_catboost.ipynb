# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
!pip install catboost
from catboost import CatBoostClassifier, Pool, cv# YOUR CODE
```

**Baseline catboost:**

Делим на train и valid:

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(data, y, train_size=0.8)
```

Coздаем train и valid датасеты:

```python
train_data = Pool(data=X_train, label=y_train, cat_features=cat_cols)
valid_data = Pool(data=X_valid, label=y_valid, cat_features=cat_cols)
```

Обучаем модель:

```python
catboost = CatBoostClassifier(loss_function='Logloss', verbose=False)
catboost.fit(train_data)
```

Предсказываем вероятности для train и valid:

```python
y_pred_train_base = catboost.predict_proba(train_data)[:, 1]
y_pred_valid_base = catboost.predict_proba(valid_data)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score

train_auc_base = roc_auc_score(y_train, y_pred_train_base)
valid_auc_base = roc_auc_score(y_valid, y_pred_valid_base)
train_auc_base, valid_auc_base
```

Выбор оптимальных параметров(обучается примерно 15 мин):

```python
iterations = [100, 300, 500, 700, 1000]
lrs = [0.0001, 0.001, 0.01, 0.1, 1]

boosting_metrics = []
for iter in iterations:
    for lr in lrs:
        params = {
            'iterations': iter,
            'learning_rate': lr,
            'loss_function': 'Logloss',
            'custom_loss': 'AUC',
            'verbose': False
        }

        scores = cv(
            params=params,
            pool=train_data,
            fold_count=5
        )

        boosting_metrics.append([iter, lr, np.mean(scores['test-AUC-mean'][iter-1])])
```

Вывод лучших парамеров:

```python
best = 0
iter = 0
lr = 0

for i in range(len(boosting_metrics)):
    if boosting_metrics[i][2] > best:
        best = boosting_metrics[i][2]
        iter = boosting_metrics[i][0]
        lr = boosting_metrics[i][1]

best, iter, lr
```

**Оптимальная модель:**

```python
catboost_optimal = CatBoostClassifier(iterations=iter, learning_rate=lr, loss_function='Logloss', verbose=False)
catboost_optimal.fit(train_data)
test_data = Pool(data=X_test, cat_features=cat_cols)
```
